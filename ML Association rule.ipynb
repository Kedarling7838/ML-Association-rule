# Association Rule Technique.

***It finds the interesting association and relationship among large dataset of data items.this rule shows how frequently
    a itemset occures in a transaction.***

```python
data=[['milk','bread','rice','book'],
      ['bread','jam','book','pen'],
      ['jam','milk','bread','rice','eggs'],
      ['rice','eggs','pen','book'],
      ['eggs','pen','milk','bread','jam'],
      ['eggs','rice','bread','jam']]
```

```python
from mlxtend.preprocessing import TransactionEncoder
te= TransactionEncoder()
te_array=te.fit_transform(data)
te_array
```

```python
import pandas as pd
df=pd.DataFrame(te_array)
df
```

```python
df=pd.DataFrame(te_array,columns=te.columns_)
df
```

```python
df=pd.DataFrame(te_array,columns=te.columns_).astype(int)
df
```

```python
from mlxtend.frequent_patterns import apriori
itemset=apriori(df,min_support=0.5,use_colnames=True)
itemset
```

```python
from mlxtend.frequent_patterns import association_rules
res=association_rules(itemset,metric='confidence',min_threshold=0.7)
res
```
